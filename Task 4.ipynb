```python
import pandas as pd

```

```python
df = pd.read_csv(r"C:\Users\Divija Vallapuneni\Downloads\Data_Cleaning_Sample_Dataset.csv")
print(df)
```

```python
df["Sales"].fillna(df["Sales"].mean(), inplace=True)
print(df)
```

```python
df["City"].fillna("Unknown", inplace=True)
print(df)
```

```python
df = df.drop_duplicates()
print(df)
```

```python
df["City"] = df["City"].replace({
    "hyd":"Hyderabad",
    "HYD":"Hyderabad"
})
print(df)
```

```python
df["Customer"] = df["Customer"].str.strip()
print(df)
```

```python
df["Date"] = pd.to_datetime(df["Date"])
print(df)
```

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

df = df[(df["Sales"] >= Q1-1.5*IQR) &
        (df["Sales"] <= Q3+1.5*IQR)]
print(df)
```

```python
df.to_excel("Cleaned_Data.xlsx", index=False)
print(df)
```

```python
total_sales = df["Sales"].sum()
print(df)
```

```python
average = df["Sales"].mean()
print(df)
```

```python
top = df.groupby("Product")["Sales"].sum().sort_values(ascending=False).head(10)
print(df)
```

```python
monthly = df.groupby(df["Date"].dt.month)["Sales"].sum()
print(df)
```

```python
city = df.groupby("City")["Sales"].sum()
print(df)
```
